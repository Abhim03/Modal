```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read the Excel files into dataframes
pre_game_df = pd.read_excel('préjeu.xlsx')
post_game_df = pd.read_excel('postjeu.xlsx')


print(pre_game_df.head())
print(post_game_df.head())
```

```python
# Calcule des les statistiques descriptives pour les jeux pré
pre_game_descriptive = pre_game_df.describe()


print("Statistiques Descriptives Pré-Jeu:")
print(pre_game_descriptive)
```

```python
# Calcule des les statistiques descriptives pour les jeux post

post_game_descriptive = post_game_df.describe()

print("Statistiques Descriptives Post-Jeu:")
print(post_game_descriptive)
```

Analyse de Corrélation

```python
pre_game_corr = pre_game_df.corr()

print("Matrice de Corrélation Pré-Jeu:")
print(pre_game_corr)


# Visualisation de la matrice de corrélation pour le pré-jeu
plt.figure(figsize=(10, 8))
sns.heatmap(pre_game_corr, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Matrice de Corrélation Pré-Jeu')
plt.show()
```

```python
post_game_corr = post_game_df.corr()

print("Matrice de Corrélation Post-Jeu:")
print(post_game_corr)

# Visualisation de la matrice de corrélation pour le post-jeu

plt.figure(figsize=(10, 8))
sns.heatmap(post_game_corr, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Matrice de Corrélation Post-Jeu')
plt.show()
```

```python
# Calculer la matrice de corrélation entre les données pré-jeu et post-jeu

correlations = pre_game_df.corrwith(post_game_df)

# Convertersion de la série de corrélations en DataFrame pour la visualisation

corr_df = pd.DataFrame(correlations, columns=['Correlation'])
corr_df['Question'] = corr_df.index
corr_df = corr_df.reset_index(drop=True)
```

```python
# Visualisation de la matrice de corrélation

plt.figure(figsize=(10, 8))  
sns.heatmap(corr_df[['Correlation']].T, annot=True, cmap='coolwarm', fmt=".2f",
            xticklabels=corr_df['Question'])
plt.title('Corrélation entre les réponses pré et post-jeu')
plt.xticks(rotation=45, ha='right') 
plt.yticks([])  
plt.show()
```
